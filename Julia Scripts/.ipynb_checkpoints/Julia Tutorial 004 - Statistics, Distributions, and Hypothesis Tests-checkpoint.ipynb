## Julia Tutorial 004 - Distributions and Statistics

In this tutorial we will walk through statistics in Julia.   We will be using a few different packages here: Statistics, StatsBase, and Distributions.   The names of the first two packages can be a bit misleading.   Statistics is actually included in base Julia in your initial install; as a result you do not need to separately download and install it.   StatsBase has some optional additional functionality; however it is possible in the future these functions will be ported into Statistics.   We will also use the RDatasets to load the "mpg" data and the Plots package for visualizing data.

```julia
# Load required packages into the environment
using Statistics, StatsBase, Distributions, RDatasets, Plots, Random
```

We will start by illustrating the Distributions package, and generate data from a variety of different statistical distributions.

```julia
# Set the random seed
Random.seed!(567);
```

```julia
normalData = rand(Normal(100, 10), 1000)
histogram(normalData)
```

```julia
uniformData = rand(Uniform(0, 50), 1000)
histogram(uniformData)
```

```julia
exponentialData = rand(Exponential(100), 1000)
histogram(exponentialData)
```

```julia
binomialData = rand(Binomial(30, 0.3), 1000)
histogram(binomialData)
```

```julia
poissonData = rand(Poisson(20), 1000)
histogram(poissonData)
```

```julia
fit(Normal, normalData)
```

```julia
fit(Exponential, exponentialData)
```

Now that we have seen all the statistical distributions as well as fitting the empirical distribution to a theoretical one, let's take a look at the statistical functions.

```julia
# Load the "mpg" dataset from the "ggplot2" R package
mpg = dataset("ggplot2", "mpg");
```

```julia
describe(mpg)
```

```julia
Statistics.mean(mpg["Cty"])
```

```julia
StatsBase.mean(mpg["Cty"])
```

```julia
mean(mpg["Cty"])
```

```julia
median(mpg["Cty"])
```

```julia
std(mpg["Cty"])
```

```julia
var(mpg["Cty"])
```

```julia
scatter(mpg["Cty"], mpg["Hwy"])
```

```julia
cor(mpg["Cty"], mpg["Hwy"])
```

```julia
cov(mpg["Cty"], mpg["Hwy"])
```
